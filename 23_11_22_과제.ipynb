# 크롤링 과제

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

def crawl_genie_chart():
    chart_data = []
    for page in range(1,5):
        url = f'https://www.genie.co.kr/chart/top200?ditc=D&ymd=20231122&hh=14&rtm=Y&pg={page}'
        header = {'User-Agent': 'Mozilla/5.0'}
        response = requests.get(url, headers=header)
        soup = BeautifulSoup(response.text)

        table=soup.find('table',{'class':'list-wrap'})
        titles = table.findAll('a',{'class':'title'})
        artists = table.findAll('a',{'class':'artist'})

        for i, (t, a) in enumerate(zip(titles, artists)):
            title = t.text.replace('19금\n', '').strip()
            artist = a.text.strip().split('\n')[0]
            rank = (i + 1) + (page - 1) * 50
            dic = {'순위': rank, '아티스트': artist, '제목': title}
            chart_data.append(dic)

    return chart_data
```

```python
chart_data = crawl_genie_chart()
df = pd.DataFrame(chart_data)
df
```
